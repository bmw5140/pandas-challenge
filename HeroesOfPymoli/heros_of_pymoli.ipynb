```python
import pandas as pd
```

```python
data_file_csv = "Resources/purchase_data.csv"
```

```python
df = pd.read_csv(data_file_csv)
```

```python
unique_sn_df = df.drop_duplicates(subset="SN", keep='first')

total_players = unique_sn_df["SN"].count()

total_players_dict = [{"Total Players" : total_players}]

total_players_df = pd.DataFrame(total_players_dict)

total_players_df
```

```python
unique_items_df = df.drop_duplicates(subset="Item Name", keep='first')
unique_items = unique_items_df["Item Name"].count()

number_of_purch = df["Purchase ID"].count()

total_revenue = df["Price"].sum()

avg_price = total_revenue / number_of_purch
```

```python
summary_dict = [{"Number of Unique Items" : total_players,
                 "Average Price" : unique_items,
                 "Number of Purchases" : number_of_purch,
                 "Total Revenue" : total_revenue
               }]

summary_df = pd.DataFrame(summary_dict)
summary_df
```

```python
unique_gender_count = unique_sn_df["Gender"].value_counts()
unique_gender_count_df = pd.DataFrame(unique_gender_count)

unique_gender_count_df = unique_gender_count_df.rename(columns = {"Gender" : "Total Count"})

unique_gender_count_df["Percentage of Players"] = (unique_gender_count_df["Total Count"] / total_players) * 100

unique_gender_count_df
```

```python
gender_groupby = df.groupby("Gender")
gender_group_count = gender_groupby["Gender"].count()
gender_group_df = pd.DataFrame(gender_group_count)
gender_group_df = gender_group_df.rename(columns = {"Gender" : "Purchase Count"})

avg_purch_gender = gender_groupby["Price"].mean()
gender_group_df["Average Purchase Value"] = avg_purch_gender

total_purch_gender = gender_groupby["Price"].sum()
gender_group_df["Total Purchase Value"] = total_purch_gender

gender_group_df["Avg Total per Person"] = total_purch_gender / unique_gender_count

gender_group_df
```

```python
bins = [0, 9.99, 14, 19, 24, 29,
        34, 39, 999]

group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", 
                "30-34", "35-39", "40+"]

age_cut = pd.cut(unique_sn_df["Age"], bins, labels=group_labels)
unique_sn_df["Age Bin"] = age_cut
age_bin_counts = unique_sn_df["Age Bin"].value_counts()

age_bin_counts_df = pd.DataFrame(age_bin_counts)
age_bin_counts_df = age_bin_counts_df.rename(columns = {"Age Bin" : "Total Count"})
age_bin_counts_df["Percentage of Players"] = (age_bin_counts_df["Total Count"] / total_players) * 100

age_bin_counts_df
```

```python
bins = [0, 9.99, 14, 19, 24, 29,
        34, 39, 999]

group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", 
                "30-34", "35-39", "40+"]

dup_age_cut = pd.cut(df["Age"], bins, labels=group_labels)
df["Age Ranges"] = dup_age_cut

dup_bin_groupby = df.groupby("Age Ranges")

dup_bin_groupby_count = dup_bin_groupby["Age Ranges"].count()
```

```python
dup_bin_counts_df = pd.DataFrame(dup_bin_groupby_count)
dup_bin_counts_df = dup_bin_counts_df.rename(columns = {"Age Ranges" : "Purchase Count"})
```

```python
avg_purch_by_age = dup_bin_groupby["Price"].mean()
dup_bin_counts_df["Average Purchase Price"] = avg_purch_by_age

age_total_purch = dup_bin_groupby["Price"].sum()
dup_bin_counts_df["Total Purchase Value"] = age_total_purch

dup_bin_counts_df["Avg Total per Person"] = age_total_purch / age_bin_counts

dup_bin_counts_df
```

```python
sn_grouby = df.groupby("SN")
sn_purch_count = sn_grouby["SN"].count()
sn_purch_count_df = pd.DataFrame(sn_purch_count)
sn_purch_count_df = sn_purch_count_df.rename(columns = {"SN" : "Purchase Count"})

sn_purch_count_df["Total Purchase Value"] = sn_grouby["Price"].sum()

sn_purch_count_df["Average Purchase Price"] = sn_purch_count_df["Total Purchase Value"] / sn_purch_count_df["Purchase Count"]

top_spenders = sn_purch_count_df.sort_values(by=["Total Purchase Value"], ascending=False)

top_spenders.head()
```
